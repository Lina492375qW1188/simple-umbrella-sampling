```python
import numpy as np
import matplotlib.pyplot as plt
from swim import harmonic, double_well, simple_md

np.set_printoptions(precision=5)
```

#### Take a look at our double well potential

```python
dw = double_well()
dw.param = {'B':2, 'x0':1}

x_arr = np.linspace(-1.5, 1.5, 50)
y_arr = dw.act(x_arr)

fig, ax = plt.subplots(figsize=(5,3.5), nrows=1, ncols=1)
ax.plot(x_arr, y_arr)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

```python
def replica(idx, x0):
    """
    Generate a replica of umbrella sampling simulations.
    """
    dw = double_well()
    dw.param = {'B':2., 'x0':1.}

    hm = harmonic()
    hm.param = {'K':10., 'x0':x0}

    md = simple_md(seed=45932)
    md.set_param(dt=0.01, kT=0.1, damping=10.)
    md.set_init_x(x0)

    md.external_pe = [dw, hm]
    md.run(int(1e6))
    
    np.savetxt('./DATA_wham/umbrella_{}'.format(idx), np.array(md.traj[::100])[:,:2])
```

#### Generate the centers of windows for umbrella sampling

```python
centers = np.arange(-15, 18, 3)*0.1
```

#### Umbrella simulations

```python
for i, center in enumerate(centers):
    replica(i, center)
```
