```python
import numpy as np
import matplotlib.pyplot as plt
from swim import double_well

np.set_printoptions(precision=3)
```

#### Generate double well potential

```python
dw = double_well()
dw.param = {'B':2, 'x0':1}

x_arr = np.linspace(-1.5, 1.5, 50)
y_arr = dw.act(x_arr)
```

#### load wham resuls

```python
x_umbrella, y_umbrella = np.loadtxt('./wham/wham/freefile_bin15', usecols=(0,1), unpack=True)
x_umbrella2, y_umbrella2 = np.loadtxt('./wham/wham/freefile_bin30', usecols=(0,1), unpack=True)
```

#### Check whether our wham analysis is insensitive to bin width

```python
fig, ax = plt.subplots(figsize=(12,5), nrows=1, ncols=2)

ax[0].plot(x_umbrella, y_umbrella, label='umbrella')
ax[0].plot(x_arr, y_arr, label='actual')
ax[0].set_title('Bin=15')
ax[0].set_xlabel('x',  size=12)
ax[0].set_ylabel('Free energy', size=12)
ax[0].legend()

ax[1].plot(x_umbrella2, y_umbrella2, label='umbrella')
ax[1].plot(x_arr, y_arr, label='actual')
ax[1].set_title('Bin=30')
ax[1].set_xlabel('x',  size=12)
ax[1].set_ylabel('Free energy', size=12)
ax[1].legend()

plt.gcf().subplots_adjust(left=0.06, right=0.98, bottom=0.1, top=0.95) # to make room for the label
plt.savefig('wham', dpi=300)
plt.show()
```
